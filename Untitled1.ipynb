```python
#do your imports and configuration here:
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
signals = np.loadtxt('data/signals.dat', comments = '//', delimiter = ';')
```

```python
fig, ax = plt.subplots()
x = signals[0]
ax.plot(x,signals[1], label = 'exponential sinusoid')
ax.plot(x,signals[2], label = 'chirping sinusoid')
ax.plot(x,signals[3], label = 'exponential decay')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Signal (mv)')
plt.title('Signal Over Time')
```

```python
plt.savefig('signals.png', dpi = 200)
```
